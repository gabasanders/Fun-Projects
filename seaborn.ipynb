```python
import seaborn as sns
```

# Tipos de plot

```python
tips = sns.load_dataset('tips')
sns.displot(tips['total_bill'],kde=True)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(tips,hue='sex')
```

# Usando categorias

```python
sns.barplot(x='sex', y='total_bill',data=tips) # Estima a média gasta pelos homens e pelas mulheres 
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
# Aqui, cada etapa consiste em 25% dos dados. A reta central é a mediana, e os pontinhos fora são os considerados 'outliers' por parte do plot.
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

# Regressão Linear


```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100},aspect=2.0)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='time',row='day')
```

# Plots Matriciais

```python
fligths = sns.load_dataset('flights')
df_flights = fligths.pivot_table(index='year',columns='month',values='passengers')
df_flights
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20,8))

sns.heatmap(df_flights, ax=ax,annot=True,fmt='.0f')
```
